## 1.1 Download data

```python
url = r'https://commondatastorage.googleapis.com/books1000/'
root = r'C:\Users\7153678\Desktop\AI\src\nlp\data\mnist'
last_report_percent = None
```

```python
def download_percent_hook(block_cnt, block_size, total_size):
    global last_report_percent
    cur_percent = int(block_cnt*block_size*100/total_size)
    
    if not cur_percent == last_report_percent:
        if not cur_percent%5:
            sys.stdout.write('{}%'.format(cur_percent))
            sys.stdout.flush()
        else:
            sys.stdout.write('.')
            sys.stdout.flush()
    last_report_percent = cur_percent
```

```python
def download_url_data(file_name, expect_size, force_download =False):
    web_addr = os.path.join(url,file_name)
    file_path = os.path.join(root, file_name)
    
    if not os.path.exists(file_path) or force_download:
        print('Attempt to download {} from {}'.format(file_name, url))
        raw_file, _ = urlretrieve(url+file_name, file_path, download_percent_hook)
        print('Download compelete!')
        file_stat = os.stat(raw_file)
        
        if file_stat.st_size == expect_size:
            print('Size verified pass: expecet and get {}'.format(expect_size))
        else:
            raise Exception('Size verified failed: expect {} but get {}'.format(expect_size, file_stat.st_size))
            
    return file_path    
```

## 1.2 Extract data

```python
np.random.seed(123)
class_num = 10
```

```python
def extract_data(file_path, force_flag=False):
    extract_folder = os.path.splitext(os.path.splitext(file_path)[0])[0]
    
    if not os.path.exists(extract_folder) or force_flag:
        print('Attempt to extract {}, it is gonna take a while'.format(file_path))
        tar = tarfile.open(file_path)
        sys.stdout.flush()
        tar.extractall(root)
        tar.close()
        print('Extract is done!')
        subs = os.listdir(extract_folder)
        sub_num = len([1 for item in subs if os.path.isdir(os.path.join(extract_folder, item))])
        if sub_num == class_num:
            print('Found {} folders for {} class'.format(sub_num, class_num))
        else:
            raise Exception('Found folder number{} but expect class number{}'.format(sub_num, class_num))
    print(extract_folder)
    return extract_folder
```

## 1.3 Generate dataset

```python
image_size = 28
pixel_depth = 255.0
```

```python
def load_letter(folder, min_num_image):
    images = os.listdir(folder)
    dataset = np.ndarray(shape = (len(images), image_size, image_size), dtype=float)
    
    valid_num_image = 0
    for image in images:
        image_path = os.path.join(folder, image)
        try:
            image_data = imageio.imread(image_path).astype(float)
            image_data = (image_data - pixel_depth/2)/pixel_depth #????
            
            if not image_data.shape == (image_size, image_size):
                print('Image {} data size does not match, just skip it'.format(image_path))
            else:
                dataset[valid_num_image,:,:] = image_data
                valid_num_image = valid_num_image + 1
        except (IOError, ValueError) as e:
            print('Could not read image {}, get error: {}'.format(image_path, e))
    
    if valid_num_image < min_num_image:
        raise Exception('Too many images of un-matched size:{}*{}'.format(image_size,image_size))
        
    dataset = dataset[0:valid_num_image,:,:]
    
    print('Full tensor size:{}'.format(dataset.shape))
    print('Mean: {}; Std: {}'.format(np.mean(dataset), np.std(dataset)))

    return dataset
```

```python
def pickle_data(folder, min_num_image, force_flag=False):
    dataset_names = []  
    letter_folders = os.listdir(folder) 
    
    for letter in letter_folders:
        pickle_file = os.path.join(folder, letter) + '.pcikle'
        dataset_names.append(pickle_file)
        
        if not os.path.exists(pickle_file) or force_flag:
            letter_path = os.path.join(folder, letter)
            dataset=load_letter(letter_path, min_num_image)            
            try:
                with open(pickle_file, 'wb') as f:
                    print('Pickling file {}'.format(pickle_file))
                    pickle.dump(dataset, f, protocol=pickle.HIGHEST_PROTOCOL)
            except Exception as e:
                raise Exception('Can not save data to {}'.format(letter_path))
    
    return dataset_names
```

## 1.4 Randomly check image

```python
def show_images(datasets, num_per_class):
    fig, axs = plt.subplots(nrows=num_per_class, ncols=class_num, figsize=(20, 20), subplot_kw={'xticks': [], 'yticks': []})
    
    for i, letter_pickle in enumerate(datasets):
        with open(letter_pickle, 'rb') as f:
            dataset = pickle.load(f)
            index = np.random.choice(dataset.shape[0], num_per_class)
            title = re.findall(r'(.+)\.pcikle',os.path.split(letter_pickle)[1])[0]
            for m, n in enumerate(index):
                axs[m][i].imshow(dataset[n,:,:])
                axs[m][i].set_title(title, fontsize=30)
                
    plt.tight_layout()
    plt.show()
```

## 1.5 Merge and prune data

```python
def merge_datasets(datasets, train_size, valid_size=0):
    train_data = np.ndarray((train_size, image_size, image_size), dtype=np.float32)
    train_labels  = np.ndarray(train_size, dtype=np.int32)
    
    
    if valid_size:
        valid_data = np.ndarray((valid_size, image_size, image_size), dtype=np.float32)
        valid_labels  = np.ndarray(valid_size, dtype=np.int32)
    else:
        valid_data = None
        valid_labels = None

    train_class_size = train_size//class_num
    valid_class_size = valid_size//class_num
    tv_class_size = train_class_size+valid_class_size
        
    for i, pickle_file in enumerate(datasets):
        try:
            print('Process file: {}; i: {}'.format(pickle_file, i))
            with open(pickle_file, 'rb') as f:
                raw_data = pickle.load(f)
                np.random.shuffle(raw_data)
                
                if valid_class_size:              
                    valid_data[i*valid_class_size:(i+1)*valid_class_size,:,:] = raw_data[0:valid_class_size,:,:]                
                    valid_labels[i*valid_class_size:(i+1)*valid_class_size] = i
                                
                train_data[i*train_class_size:(i+1)*train_class_size,:,:] = raw_data[valid_class_size:tv_class_size,:,:]  
                train_labels[i*train_class_size:(i+1)*train_class_size] = i
        except Exception as e:
            raise Exception('Can not porcess {}, get error: {}'.format(pickle_file, e))
    
    return valid_data, valid_labels, train_data, train_labels
```

```python
train_size = 200000
valid_size = 10000
test_size = 10000

valid_data, valid_labels, train_data, train_labels = merge_datasets(train_datasets, train_size, valid_size)
print('train data shape: {}, train label shape: {}'.format(train_data.shape, train_labels.shape))
print('valid data shape: {}, valid label shape: {}'.format(valid_data.shape, valid_labels.shape)) 

_, _, test_data, test_labels = merge_datasets(test_datasets, test_size)
print('test data shape: {}, test label shape: {}'.format(test_data.shape, test_labels.shape)) 
```

```python
def randomize_data(data, labels):
    permutation = np.random.permutation(labels.shape[0])
    return data[permutation,:,:], labels[permutation]
```

```python
valid_data, valid_labels = randomize_data(valid_data, valid_labels)
train_data, train_labels = randomize_data(train_data, train_labels)
test_data, test_labels = randomize_data(test_data, test_labels )
```

```python
def show_data(data, labels, row):
    
    image_cnt = row*class_num 
    
    fig, axs = plt.subplots(nrows=row, ncols=class_num, figsize=(20, 4), subplot_kw={'xticks': [], 'yticks': []}) 
    
    index = np.random.choice(data.shape[0], image_cnt)
    for index, ax in zip(index, axs.flat):
        ax.imshow(data[index,:,:])
        ax.set_title(chr(labels[index]+65), fontsize=30)
                
    plt.tight_layout()
    plt.show()
```

## 1.7 Find overlap images

```python
def get_overlap_num(dataset1, dataset2):
    keys1 = [hash(data.tobytes()) for data in dataset1]
    keys2 = [hash(data.tobytes()) for data in dataset2]
    
    keys1 = set(keys1)
    keys2 = set(keys2)

    overlaps = set.intersection(keys1, keys2)  
#     print(overlaps)
    num = len(overlaps)
    
    return num
```

```python
from sklearn.metrics import accuracy_score, recall_score,precision_score, f1_score, roc_auc_score
from sklearn.model_selection import GridSearchCV

train_data_series = train_data.reshape(train_data.shape[0], -1)
test_data_series = test_data.reshape(test_data.shape[0], -1)
valid_data_series = valid_data.reshape(valid_data.shape[0], -1)

# parameters = {'solver':('saga', 'lbfgs'), 'C':[1, 10]}
# clf = GridSearchCV(lr, parameters, cv=5)
```

```python
def get_5_mtrix(y_true, y_pred):
    var_accuracy = accuracy_score(y_true, y_pred)
    var_recall = recall_score(y_true, y_pred,average='micro')
    var_precision = precision_score(y_true, y_pred,average='micro')
    var_f1 = f1_score(y_true, y_pred,average='micro')
    return var_accuracy, var_recall, var_precision, var_f1
```

```python
train_sizes = [50, 100, 1000, 5000]

score_array = np.ndarray((len(train_sizes),4), dtype=np.float32)

for i, size in enumerate(train_sizes):
    print('train with size: {}'.format(size))
    indexs = np.random.choice(train_data.shape[0], size)
    
    x_train = train_data_series[indexs,:]
    y_train = train_labels[indexs]
  
    lr = LogisticRegression(random_state=10, multi_class='multinomial',solver='lbfgs')
    lr.fit(x_train, y_train)
        
    test_pred = lr.predict(test_data_series)
    score_array[i,:] = get_5_mtrix(test_labels, test_pred)
    
df = pd.DataFrame(score_array, columns=['accuracy', 'recall', 'precision', 'f1'])
df.insert(0, 'train size', train_sizes)
    
```
