```python
import jieba
import gensim
from gensim.models import Word2Vec
from gensim.test.utils import datapath
from gensim.models.word2vec import LineSentence
import pickle
import re
import os
import numpy as np
import opencc
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt 
```

```python
pwd = os.getcwd()
raw_dir = os.path.join(pwd, 'wiki_ch_text')
word_dir = 'wiki_ch_word'
```

```python
sum([len(os.listdir(os.path.join(raw_dir,sub))) for sub in os.listdir(raw_dir)])
```

```python
def get_all_files2(dir):
    if os.path.isfile(dir):
        return [dir]
    else:
        res= []
        subs = os.listdir(dir)
        for sub in subs:
            sub_dir = os.path.join(dir, sub)
            res = res + (get_all_files2(sub_dir))
        return res
```

```python
%%time
res2 = get_all_files2(raw_dir)
for i, raw_file in enumerate(res2):
    if i % 30 == 0:
        print(raw_file)
    with open(raw_file, 'r', encoding='utf-8') as rf:
        contents = rf.read()
    rf.close()
    sentences = contents.splitlines()
    new_file = raw_file.replace('wiki_ch_text',word_dir)
    up_dir = os.path.split(new_file)[0]
    if not os.path.exists(up_dir):
        os.makedirs(up_dir)
    with open(new_file, 'a', encoding='utf-8') as wf:
        for s in sentences:
            words = [s for s in jieba.cut(s) if not re.findall(r'\W', s)]  
            if words:
                wf.write(' '.join(words)+ '\n')
    wf.close()
```

```python
def is_all_chinese(s): 
    pattern = re.compile(r'[^\u4E00-\u9FA5]')
    if pattern.findall(s):
        return False
    else:
        return True
```

```python
def tsne_plot(model, n):
    labels = []
    tokens = []
    
    all_words = list(model.wv.vocab.keys())
    words = all_words[0:n]
    
    for word in words:
        if is_all_chinese(word):
            tokens.append(model[word])
            labels.append(word)
        
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)
    
    data = np.array(new_values)
    x = data[:,0]
    y = data[:,1]
    
    plt.figure(figsize=(20, 20))
    plt.scatter(x, y)
    for i, label in enumerate(labels):
        plt.annotate(label, xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha = 'right',
                     va = 'bottom')
    plt.show() 
```
